Find maximimum infection rate for China,Italy and Spain

```python
df_covid_aggragated.loc['China'].diff().max()
```

Find Maximun infeccion rate for all countries

```python
countries = list(df_covid_aggragated.index)
max_infection_rate=[]
for c in countries:
    max_infection_rate.append(df_covid_aggragated.loc[c].diff().max())
df_covid_aggragated['max_infection_rate']=max_infection_rate
```

```python
df_covid_aggragated.head()
    
```

Create a new df with only needed column

```python
covid_data=pd.DataFrame(df_covid_aggragated['max_infection_rate'])
```

```python
covid_data.head()
```
